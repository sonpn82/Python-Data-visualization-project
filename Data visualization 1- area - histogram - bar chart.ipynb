```python
import numpy as np
import pandas as pd
```

```python
df_can = pd.read_excel('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2
                      )
print('Data downloaded and read into a dataframe!')
```

```python
df_can.head()
```

```python
print(df_can.shape)
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df_can.head()
```

```python
df_can.drop(['Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47','Unnamed: 48','Unnamed: 49','Unnamed: 50'], axis=1, inplace=True)
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.head()
```

```python
all(isinstance(column, str) for column in df_can.columns)
```

```python
df_can.columns = list(map(str, df_can.columns)) # Change all columns to str
all(isinstance(column, str) for column in df_can.columns)
```

```python
df_can.set_index('Country', inplace=True) # Select Country as index
df_can.head()
```

```python
df_can['Total'] = df_can.sum(axis=1) # Add total column
df_can.head()
```

```python
df_can.shape
```

```python
years = list(map(str, range(1980, 2014))) # create a year list
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # for ggplot-like style
print('Matplotlib version: ', mpl.__version__)
```

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True) # sort by column total
df_top5 = df_can.head() # get top 5 rows
df_top5 = df_top5[years].transpose() # transpose to move years to row
df_top5.head()
```

# AREA PLOT

```python
df_top5.index = df_top5.index.map(int) # change index to int type
df_top5.plot(kind='area',
            alpha=0.25,       # transparent, default=0.5      
            stacked=False,    
            figsize=(20, 10),
            )
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
ax = df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

# Histogram

```python
df_can['2013'].head()
```

```python
count, bin_edges = np.histogram(df_can['2013'])
print(count) # frequency count
print(bin_edges) # bin range, default = 10 bins
```

```python
df_can['2013'].plot(kind='hist', figsize=(8 ,5), xticks=bin_edges) # fit x axis label with bin size using xticks

plt.title('Histogram of Immigration from 195 Countries in 2013') # add a title to the histogram
plt.ylabel('Number of Countries') # add y-label
plt.xlabel('Number of Immigrants') # add x-label

plt.show()
```

```python
df_can.loc[['Denmark','Norway','Sweden'], years]
```

```python
df_t = df_can.loc[['Denmark','Norway','Sweden'], years].transpose()
df_t.head()
```

```python
df_t.plot(kind='hist', figsize=(10, 6))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10 # add a buffer of 10 at start
xmax = bin_edges[-1] + 10 # add buffer of 10 at last position

df_t.plot(kind = 'hist',
        figsize=(10, 6),
        bins=15,
        stacked=True,
        alpha=0.6,
        xticks=bin_edges,
        color=['coral', 'darkslateblue', 'mediumseagreen'],
        xlim=(xmin,xmax)
        )
plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

# Bar Charts

```python
df_iceland = df_can.loc['Iceland', years]
df_iceland.head()
```

```python
df_iceland.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of immigrants') # add y-label to the plot
plt.title('Icelandic immigrants to Canada from 1980 to 2013') # add title to the plot

plt.show()
```

```python
# Add annotation to above chart
df_iceland.plot(kind='bar', figsize=(10,6), rot=90)

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

plt.annotate('',                # blank for no text
            xy=(32, 70),        # head of arrow location
            xytext=(28, 20),    # base of arrow location
            xycoords='data',    # use the coordinate system of the object
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue',lw=2)
            )
plt.annotate('2008 - 2011 Financial Crisis', # text to display
             xy=(28, 30),                    # start the text at at point (year 2008 , pop 30)
             rotation=72.5,                  # based on trial and error to match the arrow
             va='bottom',                    # want the text to be vertically 'bottom' aligned
             ha='left',                      # want the text to be horizontally 'left' algned.
            )
plt.show()
```

```python
df_can.sort_values(by='Total', ascending=True,inplace=True)
df_top15 = df_can['Total'].tail(15)
df_top15
```

# Horizontal bar chart

```python
df_top15.plot(kind='barh', figsize=(12,12), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')

# annotate value labels to each country
for index, value in enumerate(df_top15):
    label = format(int(value), ',') # format int with commas
    plt.annotate(label, xy=(value-47000, index-0.1), color='white')
plt.show()
```
